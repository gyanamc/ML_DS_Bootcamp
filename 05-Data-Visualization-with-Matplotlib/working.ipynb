```python
import numpy as np
x=np.linspace(0,5,11)
y = x**2
```

```python
#00

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)

axes.plot(x,y)
axes.set_xlabel('Xlabel')
axes.set_ylabel('Ylabel')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.0,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols =2)
#axes.plot(x,y)
plt.tight_layout()

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1, ncols =2)

axes[0].plot(x,y)
axes[0].set_title('first plot')

axes[1].plot(y,x)
axes[1].set_title('second plot')
```

## figure size and DPI

```python
fig,axes = plt.figure(nrows=2, ncols = 1,figsize=(,2))
axes.plot(x,y)

ax =fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols = 1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
#axes.plot(x,y)
plt.tight_layout()
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x square')
ax.plot(x,x**3, label='x cube')

ax.legend(loc=0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',linewidth=13,alpha = 0.5, linestyle='--',marker ='+', 
        markersize=20,markerfacecolor='black')#RGB hex code
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
